```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CampeaoCblol_2017_2024 - PÃ¡gina1.csv')
df
```

```python
df_copy = df.copy()
df_copy.columns
```

```python
df_copy.info()
```

```python
df_copy = df_copy.drop(['Name', 'Season', 'Region','Games'],axis=1)
df_copy
```

```python
df_copy = df_copy.drop(['Game duration'], axis=1)
df_copy['Win rate'] = df_copy['Win rate'].str.rstrip('%').astype('float64')
df_copy['PPG'] = df_copy['PPG'].replace('-', 0).astype('float64')
df_copy['VGPG'] = df_copy['VGPG'].replace('-', 0).astype('float64')
```

```python
df_copy.info()
```

```python
# Escalar os dados
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df_copy)

# Aplicar PCA para reduzir para 2 componentes principais
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

copy = df_copy.copy()

# Adicionar os resultados do PCA e clusters ao DataFrame original
copy['PCA1'] = data_pca[:, 0]
copy['PCA2'] = data_pca[:, 1]

# Plotar os clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(x='PCA1', y='PCA2', hue='Winner', palette='viridis', data=copy, s=100)
plt.title('Clusters de Jogadores com PCA')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay
from tensorflow.keras import Sequential
from tensorflow.keras.layers import InputLayer, Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import tensorflow as tf
```

```python
# Supondo que seu DataFrame se chama 'df' e que a coluna alvo Ã© 'target'
X = df_copy.drop('Winner', axis=1)  # Features
y = df_copy['Winner']  # Target


# Normalizando os dados
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Converter X e y para arrays NumPy (se nÃ£o forem)
X = np.array(X)
y = np.array(y)

# Definir o nÃºmero de folds
k = 5

# Inicializar o KFold
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Inicializar uma lista para armazenar as pontuaÃ§Ãµes
scores = []

# Loop sobre cada fold
for train_index, test_index in kf.split(X):
    # Dividir os dados em conjunto de treino e teste
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Criar o modelo
    model = Sequential([
        Dense(300, activation='relu', input_shape=(X_train.shape[1],)),
        Dense(200, activation='relu'),
        Dense(100, activation='relu'),
        Dense(2, activation='softmax')  # NÃºmero de classes no output layer
    ])

    # Compilar o modelo
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    # Treinar o modelo
    model.fit(X_train, y_train, epochs=200, verbose=0)

    # Avaliar o modelo
    loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
    scores.append(accuracy)

    print(f"Fold {len(scores)} - AcurÃ¡cia: {accuracy:.4f}")

# Exibir a mÃ©dia e o desvio padrÃ£o das acurÃ¡cias
print(f"\nMÃ©dia da acurÃ¡cia: {np.mean(scores):.4f}")
print(f"Desvio padrÃ£o da acurÃ¡cia: {np.std(scores):.4f}")
```
